### Google Trends unofficial API

```python
from pytrends.request import TrendReq
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.style.use(['dark_background', 'seaborn-talk'])
```

Define keywords

```python
keywords = ['World Cup', 'France']
```

Compose a request

```python
pytrend = TrendReq(hl='en-US')
pytrend.build_payload(kw_list=keywords, timeframe='2018-05-01 2018-07-20', geo='', gprop='')
```

Save the request as a dataframe

```python
interest_over_time_df = pytrend.interest_over_time()
interest_over_time_df.shape
```

Plot

```python
interest_over_time_df.plot()
plt.show()
```
